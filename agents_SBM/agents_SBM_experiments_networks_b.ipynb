```python
import evolving_SBM_generate
import matplotlib.pyplot as plt
```

### Network0_b

```python
# Example usage:
t_start = 0
delta1 = 150
t_end = 2 * delta1

num_cores = 4

block_mod_func = evolving_SBM_generate.make_step_block_probs(delta1,  num_cores=num_cores, basis_num_communities = 2, powers_num_communities = [2, 1], list_p_within_community=[49/50] * 2)
net = evolving_SBM_generate.generate_evolving_SBM(block_mod_func=block_mod_func, inter_tau= 5, activ_tau= 5, n_per_group=10, n_groups = num_cores, t_start= t_start, t_end=t_end)
```

```python
plt.matshow(net.compute_static_adjacency_matrix(start_time=0, end_time=150).toarray())
#plt.savefig('/home/b/skoove/Desktop/growing300/aggregating1.png', format='png', dpi=300)
plt.matshow(net.compute_static_adjacency_matrix(start_time=150, end_time=300).toarray())
```
