```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate 2 2d grids for the x & y bounds
p_array = np.linspace(0, 1, 200)
N_array = range(2, 10)

z = np.zeros((len(p_array),len(N_array)))

for i, p in enumerate(p_array):
    for j, N in enumerate(N_array):
        sample_X = np.random.binomial(size=200, n=1, p= p)
        sample_Y = np.random.binomial(size=200, n=N, p= p)
        z[i,j] = np.mean(np.divide(sample_X, sample_Y+1))
```

```python
# # x and y are bounds, so z should be the value *inside* those bounds.
# # Therefore, remove the last value from the z array.
# z = z[:-1, :-1]
z_min, z_max = 0, np.max(z)
levels = np.linspace(z_min, z_max, 20)

# plot
fig, ax = plt.subplots()

CC, NN = np.meshgrid(N_array, p_array)
c = ax.contourf(N_array, p_array, z, levels=levels)
fig.colorbar(c, ax=ax)

plt.show()
```
