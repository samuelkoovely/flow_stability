```python
lambda1 = 1

# Parameters
interarrival_times = [0, 2.1, 1.25, 1.5, 3, 2]
arrival_times = np.cumsum(interarrival_times)
activation_times = [1.85, 2.5, 1.5, 1.5, 1.75]
colors = ['blue', 'red', 'cyan', 'purple', 'orange']

vertices_coord = [1,2,3, 8,9,10]
events = [[1,3], [8,9], [9,10], [2,9], [1,2]]
```

```python

# Create the plot
plt.figure(figsize=(8, 4))

# Plot the vertical dashed lines and labels for t1, t2, etc.
for i, t in enumerate(arrival_times[1:]):
    # Vertical dashed lines
    plt.axvline(t, color='black', linestyle='dashed', linewidth=1)
    # Time point labels on the x-axis
    plt.text(t, -0.5, f's{i+1}', color='black', ha='center', va='bottom', fontsize=10)

# Add 0 to x axis
plt.text(0, -0.5, 0, color='black', ha='center', va='bottom', fontsize=10)

# Add <-> arrows between t1, t2, etc., and labels for interarrival times
for i in range(1, len(arrival_times)):
    plt.annotate(
        '', xy=(arrival_times[i], 0), xytext=(arrival_times[i - 1], 0),
        arrowprops=dict(arrowstyle='->', color=colors[i-1], lw=1.0)
    )
    midpoint = (arrival_times[i] + arrival_times[i - 1]) / 2
    plt.text(midpoint, +0.5, r'~Exp($\lambda_1$)', color=colors[i-1], ha='center', fontsize=10)

# Formatting the plot
plt.title('Starting Times: Poisson Process with rate ' + r'$\lambda_1$')
plt.xticks([])  # Remove default ticks on x-axis
plt.yticks([])  # Remove y-axis ticks
plt.xlim(0, arrival_times[-1] + 2 * activation_times[-1])
plt.ylim(0, len(events)+2)


plt.box(False)

plt.tight_layout()
# Show the plot
plt.show()
```

```python
# Create the plot
plt.figure(figsize=(8, 4))

# Plot the vertical dashed lines and labels for t1, t2, etc.
for i, t in enumerate(arrival_times[1:]):
    # Vertical dashed lines
    plt.axvline(t, color=colors[i], linestyle='dashed', linewidth=1)
    # Time point labels on the x-axis
    plt.text(t, -0.5, f's{i+1}', color='black', ha='center', va='bottom', fontsize=10)

# Plot the vertical dashed lines and labels for t1, t2, etc.
for i, t in enumerate(arrival_times[1:]):
    # Vertical dashed lines
    plt.axvline(t + activation_times[i], color=colors[i], linestyle='dashed', linewidth=1)
    # Time point labels on the x-axis
    plt.text(t + activation_times[i], -0.5, f'e{i+1}', color='black', ha='center', va='bottom', fontsize=10)

# Add <-> arrows between t1, t2, etc., and labels for interarrival times
for i in range(1, len(arrival_times)):
    plt.annotate(
        '', xy=(arrival_times[i], i), xytext=(arrival_times[i]+activation_times[i-1], i),
        arrowprops=dict(arrowstyle='<-', color=colors[i-1], lw=1.0)
    )
    midpoint = (2 * arrival_times[i] + activation_times[i-1]) / 2
    plt.text(midpoint, i+0.25, r'~Exp($\lambda_2$)', color=colors[i-1], ha='center', fontsize=10)

# Formatting the plot
plt.title('Ending Times ' + r'~Exp($\lambda_2$)')
plt.xticks([])  # Remove default ticks on x-axis
plt.xlabel('Time', loc='right')
plt.yticks([])  # Remove y-axis ticks
plt.xlim(0, arrival_times[-1] + 2* activation_times[-1])
plt.ylim(0, len(events)+2)
plt.box(False)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create the plot
plt.figure(figsize=(8, 4))

# Plot vertical dashed lines, labels, and shaded areas for t1, t2, etc.
for i, t in enumerate(arrival_times[1:]):
    # First vertical dashed line (start point)
    plt.text(t, -0.5, f's{i+1}', color='black', ha='center', va='bottom', fontsize=10)
    
    # Second vertical dashed line (end point)
    end_time = t + activation_times[i]
    plt.text(end_time, -0.5, f'e{i+1}', color='black', ha='center', va='bottom', fontsize=10)
    
    # Shaded area between the two lines
    plt.fill_betweenx(events[i], t, end_time, color=colors[i], alpha=0.3)

# Adding Nodes

for i in range(1,7):
    plt.text(0.15, i, f'v{7-i}', color='black', ha='center', va='bottom', fontsize=10)
    
# Formatting the plot
plt.title('Resulting Dynamic Network')
plt.xticks([])  # Remove default ticks on x-axis
plt.yticks([])  # Remove y-axis ticks
plt.xlim(0, arrival_times[-1] + 2 * activation_times[-1])
plt.ylim(0, len(events)+2)
plt.box(False)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create the 3x1 subplot
fig, axs = plt.subplots(3, 1, figsize=(10, 8))

# First plot: Starting Times
for i, t in enumerate(arrival_times[1:]):
    axs[0].axvline(t, color=colors[i], linestyle='dashed', linewidth=1)
    axs[0].text(t, -0.75, f's{i+1}', color='black', ha='center', va='bottom', fontsize=10)
axs[0].text(0, -0.75, 0, color='black', ha='center', va='bottom', fontsize=10)

for i in range(1, len(arrival_times)):
    axs[0].annotate('', xy=(arrival_times[i], 0), xytext=(arrival_times[i - 1], 0),
                    arrowprops=dict(arrowstyle='->', color=colors[i-1], lw=1.0))
    midpoint = (arrival_times[i] + arrival_times[i - 1]) / 2
    axs[0].text(midpoint, +0.5, r'~Exp($\lambda_1$)', color=colors[i-1], ha='center', fontsize=10)

axs[0].set_xticks([])
axs[0].set_xlabel('Time', loc='right')
axs[0].set_yticks([])
axs[0].set_xlim(0, arrival_times[-1] + 2 * activation_times[-1])
axs[0].set_ylim(0, len(arrival_times)+1)
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)
axs[0].spines['left'].set_visible(False)

# Second plot: Ending Times
for i, t in enumerate(arrival_times[1:]):
    axs[1].axvline(t, color=colors[i], linestyle='dashed', linewidth=1)
    axs[1].text(t, -0.75, f's{i+1}', color='black', ha='center', va='bottom', fontsize=10)
    end_time = t + activation_times[i]
    axs[1].axvline(end_time, color=colors[i], linestyle='dashed', linewidth=1)
    axs[1].text(end_time, -0.75, f'e{i+1}', color='black', ha='center', va='bottom', fontsize=10)
    axs[1].annotate('', xy=(arrival_times[i+1], i+1), xytext=(arrival_times[i+1] + activation_times[i], i+1),
                    arrowprops=dict(arrowstyle='<-', color=colors[i], lw=1.0))
    midpoint = (2 * arrival_times[i+1] + activation_times[i]) / 2
    axs[1].text(midpoint, i+1.25, r'~Exp($\lambda_2$)', color=colors[i], ha='center', fontsize=10)

axs[1].text(0, -0.75, 0, color='black', ha='center', va='bottom', fontsize=10)
axs[1].set_xticks([])
axs[1].set_xlabel('Time', loc='right')
axs[1].set_yticks([])
axs[1].set_xlim(0, arrival_times[-1] + 2 * activation_times[-1])
axs[1].set_ylim(0, len(arrival_times)+1)
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
axs[1].spines['left'].set_visible(False)


# Third plot: Resulting Dynamic Network
for i, t in enumerate(arrival_times[1:]):
    end_time = t + activation_times[i]
    axs[2].text(t, -0.75, f's{i+1}', color='black', ha='center', va='bottom', fontsize=10)
    axs[2].text(end_time, -0.75, f'e{i+1}', color='black', ha='center', va='bottom', fontsize=10)
    axs[2].fill_betweenx(events[i], t, end_time, color=colors[i], alpha=0.3)

axs[2].text(0, -0.75, 0, color='black', ha='center', va='bottom', fontsize=10)
axs[2].set_xticks([])
axs[2].set_xlabel('Time', loc='right')

vertices = []
for i in range(1, 7):
    #axs[2].text(0.15, i, f'v{7-i}', color='black', ha='center', va='bottom', fontsize=10)
    vertices.append(f'v{7-i}')
axs[2].set_yticks(vertices_coord)
axs[2].set_yticklabels(vertices, minor=False, rotation=0)

axs[2].set_xlim(0, arrival_times[-1] + 2 * activation_times[-1])
axs[2].set_ylim(0, max(vertices_coord))
axs[2].spines['top'].set_visible(False)
axs[2].spines['right'].set_visible(False)
axs[2].spines['left'].set_visible(False)
axs[2].grid(axis='y', alpha=0.3)


positions = [0.00] # Horizontal position for the inset
img_matrix= mpimg.imread('/home/b/skoove/Desktop/evolving_SBM/Matrix2.png')
for pos in positions:
    inset_ax = inset_axes(axs[2], width="50%", height="50%", loc="lower left", 
                                    bbox_to_anchor=(pos, 0.25, 0.28, 1), bbox_transform=axs[2].transAxes)
    inset_ax.imshow(img_matrix, aspect='equal')
    #inset_ax.matshow(matrix, aspect='equal')  # Keep aspect ratio equal to preserve square shape
    inset_ax.set_xticks([])
    inset_ax.set_yticks([])

list_titles = ['(A)', '(B)', '(C)']
for j, ax in enumerate(axs):
    ax.set_title(list_titles[j], loc='left', fontsize=10)

# Adjust layout and show the plot
plt.tight_layout()
plt.savefig('/home/b/skoove/Desktop/evolving_SBM/evolving_SBM_description_fig.eps', format='eps', dpi=300, bbox_inches='tight')
plt.show()
```
